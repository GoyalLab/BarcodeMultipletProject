```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import os
from math import ceil
import sklearn.metrics as metrics
import gzip
from pathlib import Path
import pandas as pd
import json
import seaborn as sns
from matplotlib import rcParams
import gzip
# figure size in inches
rcParams['figure.figsize'] = 16,10
```

```python
scrublet_df=pd.read_csv("/projects/p31666/zzhang/doublet-bchmk/output/scrublet/stats/all_detection_rates.tsv",sep="\t")
doublet_finder_df=pd.read_csv("/projects/p31666/zzhang/doublet-bchmk/output/doublet_finder/stats/all_detection_rates.tsv",sep="\t")
```

```python
scrublet_df=scrublet_df.sort_values(by=["Expected_Doublet_Rate","ID"])
scrublet_df.head()
```

```python
doublet_finder_df=doublet_finder_df.sort_values(by=["Expected_Doublet_Rate","ID"])
doublet_finder_df.head()
```

```python
doublet_finder_true_singlet=pd.pivot_table(doublet_finder_df, \
                                           values="True_Singlet_Detection_Rate", \
                                           index="ID", \
                                           columns="Expected_Doublet_Rate")
doublet_finder_singlet_in_potential_doublet=pd.pivot_table(doublet_finder_df, \
                                                           values="Singet_in_Potential_Doublet_Detection_Rate", \
                                                           index="ID", \
                                                           columns="Expected_Doublet_Rate")
scrublet_true_singlet=pd.pivot_table(scrublet_df, \
                                           values="True_Singlet_Detection_Rate", \
                                           index="ID", \
                                           columns="Expected_Doublet_Rate")
scrublet_singlet_in_potential_doublet=pd.pivot_table(scrublet_df, \
                                                           values="Singet_in_Potential_Doublet_Detection_Rate", \
                                                           index="ID", \
                                                           columns="Expected_Doublet_Rate")
```

```python
sns.heatmap(doublet_finder_singlet_in_potential_doublet,vmin=0.5, vmax=1.0)
plt.title("Doublet Finder Singlet Rate in Poential Doublets")
```

```python
sns.heatmap(doublet_finder_true_singlet, vmin=0.5, vmax=1.0)
plt.title("Doublet Finder True Singlet Rate")
```

```python
sns.heatmap(scrublet_singlet_in_potential_doublet, vmin=0.5, vmax=1.0)
plt.title("Scrublet Singlet Rate in Potential Doublets")
```

```python
sns.heatmap(scrublet_true_singlet, vmin=0.5, vmax=1.0)
plt.title("Scrublet True Singlet Rate")
```
